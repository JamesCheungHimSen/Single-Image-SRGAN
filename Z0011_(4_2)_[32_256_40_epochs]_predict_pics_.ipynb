```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
import os

train_dir = "/content/drive/MyDrive/Xccelerate/SRGAN" 

for img in os.listdir( train_dir + "/_test"):
  img_array = mpimg.imread(train_dir + "/_test/" + img)
  img_array = resize(img_array, (128,128))
  lr_img_array = resize(img_array,(32,32))
  mpimg.imsave(train_dir+ "/_predict_hr/" + img, img_array)
  mpimg.imsave(train_dir+ "/_predict_lr/"+ img, lr_img_array)
```
